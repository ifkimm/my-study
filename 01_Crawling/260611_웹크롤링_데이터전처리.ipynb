```python
# 회차 입력값 받은 후 로또번호

import requests
from bs4 import BeautifulSoup

round_num =input('회차번호를 입력하시오:')

url="https://www.fullayer.com/lottowinnumberdetail/fo/lottowinnumberdetailview/"+round_num
res = requests.get(url)
print("응답코드: ", res.status_code)

soup = BeautifulSoup(res.text, "lxml")

div_strongs = soup.find_all("div", attrs={"class":"circleNumber"})

# 수집한 데이터(문자열)들을 순서대로 저장할 빈 리스트 개설
numbers = []

#print(div_strongs)

# num 안에 있는 <strong> 태그를 찾고, 
    #    그 <strong> 태그가 감싸고 있는 순수한 '글자'만 .get_text()로 추출한 뒤,
    #    .append() 기능을 사용해 1번에서 만든 numbers 맨 뒤에 추가

for num in div_strongs :
    # print(num.strong.get_text())
    numbers.append(num.strong.get_text())

# print(numbers)


# 1. f-string 형식을 사용하여 중괄호{} 내에 파이썬 변수 및 연산 결과를 매핑한다.
# 2. round_num: 현재 루프의 회차를 나타내는 정수형(int) 변수값이 대입된다.
# 3. ', '.join(numbers): 문자열 요소를 가진 numbers 리스트를 콤마와 공백(', ')을 구분자로 사용하여 하나의 문자열로 결합한다.
line = f"{round_num}, {', '.join(numbers)}"
# line은 문자열

print(line)

# 파일 저장
# 새 파일 : w / 이어서 쓰기 a
with open('lotto.txt', 'a', encoding='utf-8') as f:
    f.write(line + '\n')
```

```python
# 영화 순위 정보 - 이미지 추출

import requests
from bs4 import BeautifulSoup

url="https://search.daum.net/search?w=tot&q=2017%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"
res = requests.get(url)
print("응답코드: ", res.status_code)

soup = BeautifulSoup(res.text, "lxml")

# print(soup.title)

# 이미지 추출 (c-thumb 태그 중 slot 속성이 thumb인 것 검색)
imgs = soup.find_all('c-thumb', attrs = {'slot':'thumb'})

# enumerate() 함수를 통해 imgs 리스트의 인덱스(idx)와 개별 img 태그 객체를 동시에 추출하며 반복 실행
for idx, img in enumerate(imgs):
    # 변수에 실제 URL 문자열이 저장되도록 변경
    img_url = img['data-original-src']
    img_res = requests.get(img_url)   # 이미지 다운로드를 위해 링크로 접속

    # 'movie_인덱스번호.webp' 형식의 파일명을 생성하고, 쓰기 전용 이진 파일 모드('wb')로 로컬 파일 시스템에 개설(open)한다.
    with open('movie_{}.webp'.format(idx+1), 'wb') as f:
        f.write(img_res.content)


    if idx >= 5 :
        break
```

```python
# 영화 순위 정보 - 연도별

import requests
from bs4 import BeautifulSoup

for year in range (2017, 2022):
    url="https://search.daum.net/search?w=tot&q={}%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR".format(year)
    res = requests.get(url)
    print("응답코드: ", res.status_code)

    soup = BeautifulSoup(res.text, "lxml")
    
    # print(soup.title)
    
    # 이미지 추출 (c-thumb 태그 중 slot 속성이 thumb인 것 검색)
    imgs = soup.find_all('c-thumb', attrs = {'slot':'thumb'})
    
    # enumerate() 함수를 통해 imgs 리스트의 인덱스(idx)와 개별 img 태그 객체를 동시에 추출하며 반복 실행
    for idx, img in enumerate(imgs):
        
        # 변수에 실제 URL 문자열이 저장되도록 변경
        img_url = img['data-original-src']
        img_res = requests.get(img_url)   # 이미지 다운로드를 위해 링크로 접속
    
        # 'movie_인덱스번호.webp' 형식의 파일명을 생성하고, 쓰기 전용 이진 파일 모드('wb')로 로컬 파일 시스템에 개설(open)한다.
        with open('movie_{}_{}.webp'.format(year, idx+1), 'wb') as f:
            f.write(img_res.content)
    
    
        if idx >= 5 :
            break
```

```python
# 코드 정리
# CSV 라이브러리를 활용한 구조화된 텍스트 파일 생성 및 데이터 레코드 기록 프로세스

# 1. CSV 포맷 파일의 읽기/쓰기 기능을 제공하는 파이썬 내장 csv 모듈을 불러온다(Import).
import csv

# 2. 'users.csv' 파일을 쓰기 모드('w')로 오픈한다. 
#    인코딩은 'utf-8'로 지정하여 한글 깨짐을 방지하고, newline=''을 통해 윈도우 환경 등에서 줄바꿈이 이중으로 발생하는 현상을 차단한다.
#    해당 파일 객체는 'f'라는 변수로 제어하며, with 블록이 종료되면 파일은 자동으로 닫힌다(Close).
with open('users.csv', 'w', encoding='utf-8', newline='') as f:
    
    # 3. 오픈한 파일 객체(f)를 csv.writer()에 전달하여 CSV 전용 파일 기록 객체(writer)를 생성한다.
    writer = csv.writer(f)

    # 4. .writerow() 메서드에 리스트를 전달하여 CSV 파일의 최상단 행에 컬럼명을 정의하는 헤더 레코드를 기록한다.
    writer.writerow(['회원번호', '이름', '이메일'])
    
    # 5. 첫 번째 회원의 식별자, 성명, 이메일 주소 문자열 데이터를 순서대로 매핑하여 하나의 데이터 행으로 기록한다.
    writer.writerow(['user001', '홍길동', 'hong@korea.com'])
    
    # 6. 두 번째 회원의 식별자, 성명, 이메일 주소 문자열 데이터를 순서대로 매핑하여 그 다음 데이터 행으로 기록한다.
    writer.writerow(['user002', '손오공', 'son@korea.com'])

# 7. 표준 출력 창에 문자열을 출력하여 파일 쓰기 프로세스가 정상적으로 완료되었음을 인지시킨다.
print('csv 저장완료')
```

```python
# 네이버 - 국내증시 - 시가총액
# 데이터 정제


import requests
from bs4 import BeautifulSoup

# 페이지 정보
page = 1
url="https://finance.naver.com/sise/nxt_sise_market_sum.naver?&page="+ str(page)
# page 숫자니까 문자열로 바꿔줌

res = requests.get(url)
print("응답코드: ", res.status_code)
soup = BeautifulSoup(res.text, "lxml")

# print(soup.title)

import csv

file_name = '시가총액1-50.csv'
f = open(file_name, 'w', encoding='utf-8', newline='')
writer = csv.writer(f)

title = 'N	종목명	현재가	전일비	등락률	거래량	거래대금	매수호가	매도호가	시가총액'.split('\t')
writer.writerow(title)
                                                                    
data_rows = soup.find('table', attrs = {'class':'type_2'}).find('tbody').find_all('tr')

#print(data_rows[3])

for row in data_rows:
    columns = row.find_all('td')
    if len(columns) <= 1:
        continue # 그 행은 처리하지 말고 다음 행으로 넘어가라
    # print(row)

    # 리스트 컴프리헨션
    items = [column.get_text().strip() for column in columns]
    datas = [item.replace('상승', '').replace('하락', '').replace('\n','').replace('\t', '') for item in items]
    
    # print(datas)
    writer.writerow(datas)

f.close()
```
