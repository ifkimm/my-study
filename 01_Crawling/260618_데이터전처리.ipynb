```python
# 함수 작성

def Low(str):   # 입력받은 문자열을 소문자로 바꾸는 함수 정의
    if pd.isna(str):   # 입력값이 결측치(NaN)인지 확인
        return str     # 결측치라면 변환하지 않고 그대로 반환
    else:   # 결측치가 아니라면
        result = str.lower()   # 문자열을 소문자로 변환하여 result에 저장
        return result   # 변환된 결과 반환

df['SW특기'].apply(Low)  # 'SW특기' 열의 각 데이터마다 Low 함수를 적용
```

# 네이버 블로그 [서울 버거 맛집] 웹크롤링 데이터 약 200개 - 데이터전처리

```python
# csv로 저장된 파일 열기

import pandas as pd

df= pd.read_csv('서울_버거_맛집_200.csv')

df.head(50)
```

```python
# 가. 정규식을 사용하여 데이터 전처리
```

r'...': 파이썬에서 정규표현식을 사용할 때 쓰는 'raw string' 형식
[^...]: ^는 "이 안에 있는 것들이 아닌 것"을 찾겠다
a-zA-Z0-9가-힣: 영어 소문자, 대문자, 숫자, 한글 전체를 의미
' ' (공백): 만약 공백까지 살리고 싶다면 꼭 포함
'': 찾은 특수문자들을 빈 문자열로 치환(제거)
regex=True: 정규표현식을 사용하겠다는 설정

```python
# 가-1. 'title' 컬럼의 특수문자 제거 (정규식 활용)
    # [^a-zA-Z0-9가-힣 ] : 영어, 숫자, 한글, 공백이 아닌 모든 것을 찾음

df['title'] = df['title'].str.replace(r'[^a-zA-Z0-9가-힣 ]', '', regex=True)

df.head(10)
```

1. '2026.06', '2026-06' 등 숫자와 기호로 된 날짜 제거
df['title'] = df['title'].str.replace(r'\d{2,4}[-./]\d{1,2}', '', regex=True)

2. '2026년 6월', '2026년 06월' 등 한글이 포함된 날짜 제거
df['title'] = df['title'].str.replace(r'\d{4}년\s?\d{1,2}월', '', regex=True)

```python
# 가-2. 연도/월 등 날짜가 있는 경우 제거

df['title'] = df['title'].str.replace(r'\d{2,4}[-./]\d{1,2}', '', regex=True)
df['title'] = df['title'].str.replace(r'\d{4}년\s?\d{1,2}월', '', regex=True)
```

```python
df.iloc[100:150]
```

# 나. 표준화

```python
# 나-1. 영문자를 모두 소문자로 변경

# 'title' 컬럼의 모든 대문자를 소문자로 변환
df['title'] = df['title'].str.lower()
```

```python
# (option) 영문자 모두 제거하고 싶은 경우

# 1. 영문자 제거
df['title'] = df['title'].str.replace(r'[a-zA-Z]', '', regex=True)

# 2. 제거 후 남은 불필요한 앞뒤 공백 제거
df['title'] = df['title'].str.strip()

# 3. 만약 중간에 남은 다중 공백을 하나로 줄이고 싶다면
df['title'] = df['title'].str.replace(r'\s+', ' ', regex=True)
```

# 다. 데이터 가공

```python
# 다-1. title, content에 데이터가 없으면 삭제
```

```python
# 'title' 또는 'content' 컬럼에 NaN이 있는 행 삭제
df.dropna(subset=['title', 'content'], how='any', inplace=True)
```

```python
# 둘 다 비어있을 때만 삭제하고 싶을 때
df.dropna(subset=['title', 'content'], how='all', inplace=True)
```

```python
# 삭제된 데이터(행)가 몇 개인지 알고 싶을 때

print("삭제 전 행 개수:", len(df))
df.dropna(subset=['title', 'content'], how='all', inplace=True)
print("삭제 후 행 개수:", len(df))
```

```python
# 다-2. title 데이터가 중복인 경우

# 'title' 컬럼을 기준으로 중복된 행을 삭제 (첫 번째 항목만 남기고 나머지는 삭제)
df.drop_duplicates(subset=['title'], inplace=True)
print("삭제 후 행 개수:", len(df))
```

# 라. 기타 데이터 제거

```python
# 라-1. 서울이 아닌 지역 있으면 행 제거
    # title 컬럼에 제주, 부산, 과천, 괌, 양평이 있는지 확인

filt = df['title'].str.contains('제주')

df = df[~df['title'].str.contains('과천|제주|부산|괌|양평', na=False)]
print("삭제 후 행 개수:", len(df))

# df['포함여부'] = df['title'].str.contains('|'.join(['제주', '부산', '과천', '괌']), na=False)
```

```python
# 라-2. 불용어(stopword) 제거

# 1. 제거할 단어 리스트 정의
stop_words = ['에서', '까지', '와', '과', '의', '로', '은', '는', '이', '가']

# 2. 리스트를 '|'로 연결하여 정규표현식 패턴 생성
pattern = r'\b(' + '|'.join(stop_words) + r')\b'

# 3. 패턴 적용하여 단어 삭제
df['title'] = df['title'].str.replace(pattern, '', regex=True).str.strip()
```

```python
# 라-3. 연속된 공백 제거

df['title'] = df['title'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

# NLTK 모듈 활용

```python
import nltk

# 리소스 다운로드
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')   # 불용어

# 샘플 문장
sp_sentence = 'The quick brown fox jumped over the lazy dog.'
words = nltk.word_tokenize(sp_sentence)

for word in words:
    print(word)
```

```python
# 품사 태깅
nltk.download('averaged_perceptron_tagger_eng')
tags = nltk.pos_tag(words)

for tag in tags:
    print(tag)
```

```python
# 불용어 제거

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords') 
stop_words = set(stopwords.words('english'))
new_words = [word for word in words if word not in stop_words]

for new_word in new_words:
    print(new_word)
```
