```python
# 네이버 블로그 - 탈모

import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By   # 요소를 찾아갈 때 기준이 되는 주소(선택자) 규칙을 제공 / By 객체
import csv

keyword = '탈모'
browser = webdriver.Chrome()

file_name = '탈모검색_400.csv'
csvf = open(file_name, 'w', encoding = 'utf-8-sig', newline='')
writer = csv.writer(csvf)

# 헤더 정보 (title, link, date, content)
writer.writerow(['title', 'link', 'date', 'content'])

for i in range(1, 71):
    url = 'https://section.blog.naver.com/Search/Post.naver?pageNo={}&rangeType=ALL&orderBy=sim&keyword={}'.format(i, keyword)

    
    browser.get(url)
    time.sleep(3)   # 브라우저 페이지 이동 후, 동적 텍스트 로딩이 완료될 때까지 3초간 물리 대기
    html = browser.page_source
        
    soup = BeautifulSoup(html, "lxml")
        
    blogs = soup.find_all("div", attrs={"class":"item multi_pic"})
        
    for blog in blogs :
        title = blog.find("span", attrs={"class":"title"}).get_text()
        link = blog.find('a')['href']   # 링크 정보
        date = blog.find('span', attrs = {'class':'date'}).get_text()
        
        # 링크에 접속
        res = requests.get(link)
        html_iframe = res.text
        soup_iframe = BeautifulSoup(html_iframe, 'lxml')
        iframe_src = 'https://blog.naver.com/' + soup_iframe.find('iframe', attrs = {'id' : 'mainFrame'})['src']
        # iframe은 셀레니움 쓰지 않아도 접속 / html 가져오기 가능
        
        res = requests.get(iframe_src)
        html_src = res.text
        soup_src = BeautifulSoup(html_src, 'lxml')
        content_p = soup_src.find_all('p', attrs = {'class' : 'se-text-paragraph'})
        content = ''
        for p in content_p:
            content += p.get_text('\n', strip=True)
        
        
        # print(title, link, date, content)

        line = [] # 초기화
        line.append(title)
        line.append(link)
        line.append(date)
        line.append(content)

        # 데이터를 저장
        writer.writerow(line)

csvf.close()
print('csv 저장 완료')
```
