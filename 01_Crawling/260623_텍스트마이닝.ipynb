## DTM

```python
# dtm - 띄어쓰기 기준

# 텍스트 데이터에서 단어의 등장 빈도를 세어주는 CountVectorizer 불러오기
from sklearn.feature_extraction.text import CountVectorizer
# 데이터를 보기 쉬운 표(데이터프레임) 형태로 다루기 위해 pandas 불러오기
import pandas as pd

# 1. 분석할 대상이 되는 원본 텍스트 데이터 (총 3개의 문장)
documents = [
    "나는 아빠에게 자전거를 사달라고 말했다",
    "나는 엄마랑 아빠 중에 엄마가 더 좋다",
    "나는 오늘 친구들과 놀이터에서 그네를 탔다"
]

# 2. 벡터화 도구 초기화
# CountVectorizer 객체를 생성 (기본적으로 띄어쓰기 기준으로 단어를 분리)
vectorizer = CountVectorizer()

# 3. DTM(문서-단어 행렬) 생성
# fit: 3개의 문서를 모두 분석하여 중복을 제거한 '전체 단어 사전'을 만들기
# transform: 각 문서마다 사전의 단어들이 몇 번씩 등장하는지 숫자로 변환
dtm = vectorizer.fit_transform(documents)

# 4. 기계가 만든 행렬 데이터를 사람이 보기 쉬운 표(DataFrame) 형태로 변환
dtm_df = pd.DataFrame(
    dtm.toarray(), # 압축되어 있던 행렬 데이터를 일반적인 2차원 숫자 배열로 풀어줌
    columns = vectorizer.get_feature_names_out(), # 표의 열(가로) 이름: 만들어진 '전체 단어 목록'을 넣기
    index = ['문서1', '문서2', '문서3']
)

# 5. 최종 완성된 DTM 표 출력
print(dtm_df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
from konlpy.tag import Okt


okt = Okt()
def noun_tokenizer(text):
    # 입력받은 텍스트에서 명사만 추출하여 리스트로 반환
    return okt.nouns(text)

documents = [
    "나는 아빠에게 자전거를 사달라고 말했다",
    "나는 엄마랑 아빠 중에 엄마가 더 좋다",
    "나는 오늘 친구들과 놀이터에서 그네를 탔다"
]


vectorizer = CountVectorizer(tokenizer = noun_tokenizer, token_pattern = None)

dtm = vectorizer.fit_transform(documents)


dtm_df = pd.DataFrame(
    dtm.toarray(), 
    columns = vectorizer.get_feature_names_out(),
    index = ['문서1', '문서2', '문서3']
)


print(dtm_df)
```

```python
# dtm - 형태소 분석기 활용


# 단어 등장 빈도 계산을 위한 CountVectorizer 모듈 불러오기
from sklearn.feature_extraction.text import CountVectorizer
# 행렬 데이터를 표(데이터프레임) 형태로 다루기 위한 pandas 모듈 불러오기
import pandas as pd
# 한국어 형태소 분석을 위한 Okt 모듈 불러오기
from konlpy.tag import Okt

# 형태소 분석기 객체 생성
okt = Okt()

# 텍스트에서 명사만 추출하는 사용자 정의 토크나이저 함수 정의
def noun_tokenizer(text):
    # 입력된 텍스트에서 명사만 리스트 형태로 반환
    return okt.nouns(text)

# 분석 대상이 될 3개의 원본 텍스트 데이터 준비
documents = [
    "나는 아빠에게 자전거를 사달라고 말했다",
    "나는 엄마랑 아빠 중에 엄마가 더 좋다",
    "나는 오늘 친구들과 놀이터에서 그네를 탔다"
]

# 명사 추출 함수(noun_tokenizer)를 단어 분리 기준으로 사용하는 CountVectorizer 객체 설정
vectorizer = CountVectorizer(tokenizer = noun_tokenizer, token_pattern = None)

# 문서 데이터를 분석하여 단어 사전을 구축하고 문서-단어 행렬(DTM) 생성
dtm = vectorizer.fit_transform(documents)

# 머신러닝 결과를 사람이 읽기 쉬운 표(데이터프레임) 형식으로 변환
dtm_df = pd.DataFrame(
    dtm.toarray(), # 압축된 행렬 데이터를 일반적인 2차원 숫자 배열로 변환
    columns = vectorizer.get_feature_names_out(), # 표의 열(가로) 이름에 전체 단어 목록 지정
    index = ['문서1', '문서2', '문서3'] # 표의 행(세로) 이름에 각 문서 식별명 부여
)

# 완성된 문서-단어 행렬(DTM) 표 화면 출력
print(dtm_df)
```

## IDF

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
from konlpy.tag import Okt

docs = [
    "나는 아빠 자전거",
    "나는 엄마 아빠 엄마",
    "나는 친구 놀이터 그네"
]


vectorizer = CountVectorizer()
dtm = vectorizer.fit_transform(docs)

dtm_df = pd.DataFrame(
    dtm.toarray(), 
    columns = vectorizer.get_feature_names_out(),
    index = ['문서1', '문서2', '문서3']
)


print('==== DTM 또는 TF ====')   # TF: 특정 단어가 현재 문서 안에서 얼마나 자주 등장하는가
print(dtm_df)

df_series = (dtm_df > 0).sum(axis=0)
print('==== DF ====')   # 의미: 특정 단어가 포함된 '문서의 개수'
print(df_series)


tfidf = TfidfVectorizer(norm = None)
tfidf_matrix = tfidf.fit_transform(docs)

idf_df = pd.DataFrame({
    '단어' : tfidf.get_feature_names_out(),
    'IDF' : tfidf.idf_
})

print('==== IDF ====')   # 어떤 단어가 전체 문서(수집한 모든 데이터)에서 얼마나 희귀하게 등장하는가를 나타내는 값 (DF의 역수)
print(idf_df)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns = tfidf.get_feature_names_out(),
                        index = ['문서1', '문서2', '문서3']
                       )


print('==== TF-IDF ====')   # TF 점수와 IDF 점수를 곱해서 만든 최종 키워드 중요도 점수
print(tfidf_df)
```

## 회귀분석

```python

import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({
    '공부시간': [1.5, 1.8, 2.3, 3.3, 4.6, 6.5],
    '시험점수': [50, 65, 85, 75, 90, 98]
})

# 독립변수, 종속변수

X = df[['공부시간']]
y = df['시험점수']


model = LinearRegression()
model.fit(X,y)

print('기울기:', model.coef_[0])
print('절편:', model.intercept_)

# 예측
df['예측점수:'] = model.predict(X)

print(df)
```

## 로지스틱 회귀분석

```python
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression

df = pd.DataFrame({
    '공부시간': [1.5, 1.8, 2.3, 3.3, 4.6, 6.5],
    '통과여부': [0, 0, 1, 0, 1, 1]
})

# 독립변수, 종속변수

X = df[['공부시간']]
y = df['통과여부']

model = LogisticRegression()
model.fit(X,y)

print('기울기:', model.coef_[0])
print('절편:', model.intercept_)

# 통과확률
df['통과확률:'] = model.predict_proba(X)[:, 1]

# 예측
df['예측결과'] = model.predict(X)

print(df)
```
