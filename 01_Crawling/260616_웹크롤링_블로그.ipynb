```python
# 심화: 네이버 블로그 - 셀레니움 활용
# 키워드

import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By 
import csv

keyword = '서울 버거 맛집'

# 1. 브라우저 초기화 (반복문 바깥에서 단 한 번만 실행하여 메모리 오염 방지)
browser = webdriver.Chrome()

# 2. 저장할 CSV 파일 설정 (utf-8-sig: 엑셀에서 한글 깨짐 방지 폰트 설정)
file_name = '서울맛집_200.csv'
csvf = open(file_name, 'w', encoding='utf-8-sig', newline='')
writer = csv.writer(csvf)

# 3. CSV 상단 헤더(컬럼명) 기록
writer.writerow(['title', 'link', 'date'])

# 4. 다중 페이지 수집 루프 (1페이지부터 25페이지까지)
for i in range(1, 26): # f-string 문법을 사용하여 중괄호 자리에 페이지 번호(i)를 동적으로 삽입
    url = 'https://section.blog.naver.com/Search/Post.naver?pageNo={}&rangeType=ALL&orderBy=sim&keyword={}'.format(i, keyword)

    # 브라우저 페이지 이동 후, 동적 텍스트 로딩이 완료될 때까지 3초간 물리 대기 (NoneError 방지)
    browser.get(url)
    time.sleep(3) 
    
    # 현재 화면의 HTML 소스코드를 텍스트 상태로 획득 및 뷰티풀수프 분석 객체 변환
    html = browser.page_source
    soup = BeautifulSoup(html, "lxml")
    
    # 블로그 게시글 전체를 포함하는 개별 부모 상자(div) 리스트 수집
    blogs = soup.find_all("div", attrs={"class":"item multi_pic"})
    
    # 5. 수집된 부모 상자 내부를 순회하며 세부 데이터 추출 및 매칭 (중첩 for문)
    for blog in blogs:
        # 단일 태그 내부 텍스트 추출 및 양끝 공백 제거
        title = blog.find("span", attrs={"class":"title"}).get_text(strip=True)
        date = blog.find("span", attrs={"class":"date"}).get_text(strip=True)
        
        # a 태그 내부의 href(링크 주소) 속성값을 직접 타격하여 추출
        link = blog.find('a')['href']
        
        # 6. 행 데이터 주머니 생성 및 저장 (각 반복 시마다 초기화되어 덮어쓰기 방지)
        line = [title, link, date]
        writer.writerow(line)

# 7. 파일 안전하게 닫기 및 프로세스 종료
csvf.close()
browser.quit() # 메모리 해제를 위해 열려있는 크롬 창을 완전히 종료

print('csv 저장 및 크롤링 프로세스 완료')
```
